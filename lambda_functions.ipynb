### Lambda Functions in Python
#### Lambda functions are small anonymous (a function without a name) functions defined using the __lambda__ keyword. They can have any number of arguments but only one expression. They are commonly used for short operations or as arguments to higher-order functions.
#### Syntax:   lambda arguments:expression   
##### Ex: lambda x,y:x+y

```python
### Lambda function syntax 
lambda arguments:expression 
```

```python
# This is normal function with name 
def addition(a,b):
    return a+b

addition(5,6)
```

```python
### A function without any specific name (like addition above)
lambda a,b:a+b  # a,b are arguments. a+b is the expression 

addition=lambda a,b:a+b   # assigning the lambda function to a variable
#type(addition)
print(addition(5,6))  # calling the lambda function variable with arguments (5,6)
```

```python
def even(num):
    if num%2==0:
        return True
even(24)
```

```python
even=lambda num:num%2==0
even(12)
```

```python
### multiple arguments 

def addition(x,y,z):
    return x+y+z

addition(10,10,10)
```

```python
addition1=lambda x,y,z:x+y+z
addition1(1,2,3)
```

```python
### map() - applies a function to all items in a list 
numbers=[1,2,3,4,5]
def square(number):
    return number**2

square(2)
```

```python
# the same with map() function 
# syntax: map(func,*iterables) --> map object
# map uses lazy loading technique, means it wont initialize, so use list.
values=[5,6,7]
list(map(lambda x:x**2,values))
```
