```python
### creating a tuple with "()"
empty_tuple=()
print(empty_tuple)
print(type(empty_tuple))
```

```python
lst=list()      # this is creating an empty list with "list()" method
print(type(lst))

tup=tuple()    # this is creating an empty tuple with "tuple()" method
print(type(tup))
```

```python
numbers=tuple([10,20,30])   # list can be converted into tuple
print(numbers)

lst=list((10,20,30))     # tuple can be convered into list as well 
print(lst)
```

```python
### Tuple methods 
numbers=tuple([1,2,3,1,4,5,5,9,9,9])
print(numbers.count(1))   # return the frequency of "1"
print(numbers.index(3))   # return index of "3" 
print(numbers.count(9))  
  
```

```python
### Packing and Unpacking tuples 

# packing a tuple
packed_tuple=1,"Hello",3.14  # define the tuple and it will pack by default
print(packed_tuple)

# unpacking the tuple 
a,b,c=packed_tuple  
print(a)
print(b)
print(c)

### Unpacking with * 
numbers=(1,2,3,4,5,6) 
first,*middle,last=numbers    # this * consider all the elements till before the last element 
print(first)
print(middle)
print(last)
```

```python
### Nested tuples 

# nested list (list inside a list)
lst=[[],[],[]]   # this is nested list 
lst1=[[],[],()]  # inside list, we can have a tuple as well 

# nested tuple (tuple inside a tuple)
nested_tuple= ((1,2,3),("a","b","c"),(True,False))

# access the elements inside a tuple 
print(nested_tuple[0])
print(nested_tuple[1][2])

# iterate over nested tuple 
for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item,end=" ")
    print()
```
