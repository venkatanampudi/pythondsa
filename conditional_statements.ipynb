```python
### Conditional Statements (if,elif,else)
# Introduction to conditional statements 
# if stmt     ===> evaluates the condition and execute the block of the code if the condition is "true"
# else stmt   ===> else statment executes a block of code if the condition in the if statement is "False"
# elif stmt   ===> The elif statement allows you to check multiple conditions.It stands for "else if"
# nested conditional statements  ===> You can place if, elif, or else statements inside another if, elif, or else statements.

### if statement 
age=20

if age>=18:
    print("You are allowed to vote in the elections")
```

```python
## else statment
age=16
if age>=18:
    print("you are eligble for voting")
else:
    print("you are a minor")
```

```python
### elif 
age=17

if age<12:
    print("You are a child")
elif age<=19:
    print("You are a teenager")
else:
    print("You are an adult")
```

```python
### Nested conditional statements 

# find if the number is even, odd, or negative
num=int(input("Enter your number:"))
if num>0:
    print("The number is positive")
    if num%2==0:
        print("The number is even")
    else:
        print("The number is odd")
else:
    print("The number is Zero or negative")
    
```

```python
### Use cases 

# Determine if a year is a leap year using nested conditional statement 

year = int(input("Enter the year:"))

if year%4==0:
    if year%100==0:
        if year%400==0:
            print(year, "is a leap year")
        else:
            print(year,"is not a leap year")
    else:
        print(year,"is a leap year")
else:
    print(year,"is not a leap year")
```

```python
### Create a simple calculator 

num1=float(input("Enter first number:"))
num2=float(input("Enter second number:"))
op = input("Enter operation (+,-,*,/): ")

# perform the requested operation 

if op == "+":
    res = num1 + num2 
elif op == "-":
    res = num1 - num2
elif op == "*":
    res = num1 * num2 
elif op == "/":
    if num2 != 0:
        res = num1 / num2
    else:
        res = "Error! Division by Zero."
else:
    res = "Invalid operation"

print("Result:", res)
```
