#### Definition: A function is a block of code that performs a specific task. Functions help in organizing code, reusing code, and improving readability.


```python
### Function syntax 

def functiona_name(parameters):
    """ Docstring"""

    # function body 
    return expression

### Why functions? Lets say I create a function with below block of code.
### We can reuse the function code wherever we want (reusing the code)
num=24 
if num%2==0:
    print("The number is even")
else:
    print("The numnber is odd")
```

```python
### Create/Define a function for the above code 
def even_or_odd(num):
    """This function finds even or odd"""
    if num%2==0:
        print("The number is even")
    else:
        print("The number os odd")
```

```python
### Calling the function 
### syntax: function_name(pass the parameters)
even_or_odd(24)
```

```python
### Function with multiple parameters 

def add(a,b):
    return a+b

result=add(2,4)
print(result)
```

```python
### function without default parameters
def greet(name):
    print(f"Hello {name}...Welcome to Hyatt")

greet("Venkat")
```

```python
### function with default parameter (name=Guest)
def greet(name="Guest"):
    print(f"Hello {name}...Welcome to Hyatt")

greet()
```

```python
### Variable length arguments 

# 1. Positional arguments (these args you can give any name)
def print_numbers(*args):
    for number in args:
        print(number)

# calling print_numbers function with many arguments. we can pass any number of arguments during function calling
print_numbers(1,2,3,4,5,6,9,8,"Venkat")
```

```python
# practicing positional arguments
def print_anything(*args):
    for anything in args:
        print(anything)

print_anything("Venkat",1,2,9.99,"Python","Machine Learning","Some thing")
```

```python
### Keyword arguments (represented with **). In the kwargs, all the parameters in the form of key-value pairs

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")

print_details(name="Venkat",age=25, country="USA")
```

```python
### using both same time

def print_details(*args,**kwargs):
    for val in args:
        print(f"Positional argument : {val}")

    for key,value in kwargs.items():
        print(f"Key Word argument: {key}:{value}")
```

```python
print_details(1,2,3,4,5,name="Venkat",age=25, country="USA")
```
