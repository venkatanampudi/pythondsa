```python
### 1. Create a to do list of tasks to keep a track 

todo_list=["Buy groceries","Pay bills","Clean the house"]

# Adding tasks to todo_list 
todo_list.append("Schedule a meeting with Mr.X")
todo_list.append("Go for a jog")

# Removing a completed task 
todo_list.remove("Clean the house")

# checking if task is in the list 
if "Pay bills" in todo_list:
    print("Don't forget to pay the utiity bill")


# print the todo list 
print("To do list remaining:")
for task in todo_list:
    print(f"-{task}")
```

```python
### 2. Creating a list to store and calculate average grades for students 

# Organizing student grades 
grades = [85,92,78,90,88]

# Adding  a new grade 
grades.append(95)

# print(sum(grades))
# print(len(grades))

# Calculating the average grade 
avg_grade = sum(grades) / len(grades)
print(f"Average grade: {avg_grade:.2f}")

# Finding the highest and lowest 
highest_grade = max(grades)
lowest_grade = min(grades)

print(f"Highest grade:{highest_grade}")
print(f"Lowest grade:{lowest_grade}")
```

```python
### Use a list to collect and analyze user feedback 

# collecting user feedback
feedback=["Great","Excellent","Good","Not good"]

# Adding new feedback
feedback.append("Not happy")

# couting specific feedback 
positive_feedback_count=sum(1 for comment in feedback if "Great" in comment.lower() or "Excellent" in comment.lower())
print(positive_feedback_count)
print(f"Positive feedback count:{positive_feedback_count}")

# printing all feedback
print("User Feedback:")
for comment in feedback:
    print(f"-{comment}")
```
