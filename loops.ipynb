```python
### WHILE loop : While loop continues to execute as long as the condition is True

count = 0 

while count < 5:
    print(count)
    count += 1 

```

```python
count = 0 
while count%2==0:
    print(count)
    count = count + 1 
```

```python
### Loop control statements 
# break: break statment exists the loop prematurely.

for i in range(10):
    if i == 5:  # when i become 5, control come out of loop
        break
    print(i)    # until the loop runs, print the i

```

```python
for i in range (1,100):
    if i == 10:
        break  ## exiting this loop permanently 
    print(i)
```

```python
 ### continue : the continue statement skips the current iteration and continues with the next

 for i in range(10):
    if i%2 == 0:
        continue  ## this condition will continue to next iteration if i%2==0 (which is even number).Hence it will print only odd numbers
    print(i)

```

```python
for i in range (1,20):
    if i%2 != 0:
        continue 
    print(i)
```

```python
### pass
### The pass statement is a null operation, it does nothing 
for i in range(5):
    if i==3:
        pass
    print(i)
```

```python
### Nested loops 
## A loop inside a loop 

for i in range(3):
    for j in range(2):
        print(f"i: {i} and j: {j}")
```

```python
## Example - Calculate the sum of the first N natural numbers using a while and for loop 

n=10 
sum=0 
count=1 

while count <= n:
    sum=sum+count 
    count=count+1

print("Sum of first N natural numbers: ",sum)

```

```python
n=10 
sum=0 
count = 1 

for i in range(11):
    sum = sum + i 

print("Sum of first N natural numbers: ",sum)
```

```python
### Example - Display prime numbers between 1 and 100 (prime number = divisible by 1 and itself)
for num in range (1,101):
    if num>1:
        for i in range(2,num):
            if num%i==0:
                break
        else:
            print(num)
```
