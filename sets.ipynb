### Sets 
* Sets are built-in data type in Python used to store collection of unique elements
* They are UNORDERED, meaning elements do not follow a specific order 
* Sets DO NOT allow duplicate elements
* Sets are useful for membership tests, eliminating duplicate elements, and performing mathematical set operations like UNION, INTERSECTION, DIFFERENCE, and SYMMETRIC DIFFERENCE
* NO indexing (you can't access elements by position)
* Mutable (you can add/remove elements)
* Very fast for membership checks (in)
* Set can be defined either with set() function or with empty curley braces {}

##### In the context of Lists/Tuples/Sets, item(s) called "Element(s)"
##### In the context of Dictionary, item(s) called "item(s)"

```python
# Set membership test - Whether an element is a member of set or not ?
myset = {1,2,3,5,6,9,10}
print(3 in myset) 
print(10 in myset)
print(11 in myset)
```

```python
### Mathematical operations (SET operators)
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

# Union --> combining from both sets 
unionset = set1.union(set2)
print(unionset)

# Intersection --> return common elements 
interset = set1.intersection(set2)
print(interset)

# Intersection_update ==> Whatever the common elements with "intersection" and update the set1 with those elements
set1.intersection_update(set2)
print(set1)




```

```python
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}
# Difference 
print(set1.difference(set2)) ## --> from set1 remove common elements with set2 and return the rest of set1 elements
print(set2.difference(set1)) ## --> from set2 remove common elements with set1 and return the rest of set2 elements
```

```python
### Sets methods 
set1={1,2,3}
set2={4,5,6}

# issubset 
print(set1.issubset(set2))    # checks if set1 all the elements are there in set2

# superset 
print(set1.issuperset(set2))  # checks if set1 has all the elements of set2. Here, no
```

```python
set1={1,2}
set2={4,5,6,1,2}
print(set1.issubset(set2))
print(set1.issuperset(set2))
```
